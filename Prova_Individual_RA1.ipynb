# **1. Primeira Questão:**

$$\lim_{{x \to 0}} \frac{{1 - \cos(x)}}{{x^2}}$$

<br>
Ao fazermos substituição de $x$ pelo valor $0$, percebemos que o resultado é:

$$ \lim_{{x \to 0}} \frac{1 - \cos(0)}{0^2} $$
<br>
$$ \lim_{x \to 0} \frac{0}{0}$$

Logo, como a divisão por zero é impossível de se calcular, o próximo passo é a fatoração. Entretanto, nesse caso a fatoração não é uma opção, pois $x^2 = x . x$, logo o denominador resultaria em zero, então precisamos usar a aproximação para chegarmos a um resultado satisfatório.

Pela direita, usando $x = 0.001$:

$$ \lim_{{x \to 0^+}} \frac{1- \cos(0.001)}{0.001^2} $$

$$\lim_{{x \to 0^+}} \frac{1 - 0.9999995}{0.000001} = \frac{0.0000005}{0.000001}$$

$$\lim_{{x \to 0^+}} ≈ \frac{1}{2}$$

Pela esquerda, usando $x = -0.001$:

$$ \lim_{{x \to 0^+}} \frac{1- \cos(-0.001)}{-0.001^2} $$

$$\lim_{{x \to 0^+}} \frac{1 - 0.9999995}{0.000001} = \frac{0.0000005}{0.000001}$$

$$\lim_{{x \to 0^-}} ≈ \frac{1}{2}$$

Analisando ambos os resultados, é possível afirmar que:

<br>
$$\lim_{x \to 0} \frac{1-cos(x)}{x^2} = \frac{1}{2}$$

<br><br>
Conseguimos ver isso nitidamente analisando o gráfico feito em python usando Matplotlib e Numpy:

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return (1 - np.cos(x)) / (x**2)

def plotarGrafico():
  xp = np.linspace(-8, 8, 100)
  yp = funcao(xp)

  plt.title(r'Gráfico da função $f(x) = \frac{1 - \cos(x)}{x^2}$')
  plt.plot(xp, yp, label=r'$f(x)$')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.scatter(0, 1/2, c='red', label='$\\lim_{x \\to 0} \\frac{1 - \\cos(x)}{x^2} = \\frac{1}{2}}$')
  plt.grid()
  plt.legend()
  plt.show()

plotarGrafico()
```

# **2. Segunda Questão:**

$$\lim_{x \to 1} \frac{\sqrt{x+8} -3 }{x^2 - 1}$$

Fazendo a substituição de x para 1:

$$\lim_{x \to 1} \frac{\sqrt{9}-3}{1-1}$$

$$\lim_{x \to 1} \frac{0}{0}$$

Como a divisão por zero não pode ser calculada, iremos utilizar o método de racionalização da expressão para chegar ao resultado:

$$\lim_{x \to 1} \frac{(\sqrt{x+8} -3)(\sqrt{x+8} + 3)}{(x^2 - 1)(\sqrt{x+8} + 3)}$$

$$\lim_{x \to 1} \frac{1}{(x+1)(\sqrt{x+8}+3)}$$

<br>
Fazendo a substituição do x por 1, temos que o limite no ponto 1 é:

$$\lim_{x \to 1}\frac{1}{(1+1).(\sqrt{1+8}+3)}$$

$$\lim_{x \to 1}\frac{1}{2.(3+3)}$$

$$\lim_{x \to 1}\frac{1}{12}$$


<br>
Logo chegamos ao resultado de:
$$\lim_{x \to 1} \frac{\sqrt{x+8} -3 }{x^2 - 1} = \frac{1}{12} $$

<br><br>
Conseguimos comprovar que o valor que chegamos é satisfatório analisando o gráfico da equação feito em python:

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return ((np.sqrt(x+8) -3) / (x**2 -1))

def plotarGrafico():
  xp = np.linspace(-1.5, 1.25, 100)
  yp = funcao(xp)

  plt.title('Gráfico da função $\\frac{\\sqrt{x+8} -3 }{x^2 - 1}$')
  plt.plot(xp, yp, label=r'$f(x)$')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.scatter(1, 1/12, c='red', label='$\\lim_{x \\to 1} \\frac{\\sqrt{x+8} -3 }{x^2 - 1} = \\frac{1}{12}}$')
  plt.grid()
  plt.legend()
  plt.show()

plotarGrafico()
```

# **3. Terceira Questão:**

$$\lim_{h \to 0} \frac{(2+h)^3 -8}{h}$$

Fazemdo a substituição:

$$\lim_{h \to 0} \frac{(2+0)^3 -8}{0}$$

$$\lim_{h \to 0} \frac{8-8}{0}$$

$$\lim_{h \to 0} \frac{0}{0}$$


Como é impossível fazer divião por zero, iremos realizar a expansão do numerador usando a fórmula para o cubo da soma e simplificando o resultado.

Fórmula para o cubo da soma que será usado: $ (a+b)^3 = a^3 + 3a^3b + 3ab^3 + b^3 $

<br>
Substituindo os valores temos:
$$ (2 + h)^3 = h^3 + 3h^32 + 3h2^2 + 2^3 = h^3 + 6h^2 + 12h + 8 $$

<br>
Agora substituindo na expressão original:
$$\lim_{h \to 0} \frac{h^3 + 6h^2 + 12h}{h}$$

$$ \lim_{h \to 0}h^2 + 6h + 12 $$

<br>
Agora, ao fazer a substituição do h por 0, chegamos ao resultado:
$$ \lim_{x \to 0} (0^2 + 6.0 + 12 = 12) $$

Ou seja:
$$\lim_{h \to 0} \frac{(2+h)^3 -8}{h} = 12$$


<br><br>
Podemos confirmar o resultado fazendo a análise ao gráfico feito em python:

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return (((2+x)**3 - 8) / x)

def plotarGrafico():
  xp = np.linspace(-3, 3, 100)
  yp = funcao(xp)

  plt.title('Gráfico da função $\\frac{(2+h)^3 - 8}{h}$')
  plt.plot(xp, yp, label=r'$f(h)$')
  plt.xlabel('h')
  plt.ylabel('y')
  plt.scatter(0, 12, c='red', label='$\\lim_{h \\to 0} \\frac{(2+h)^3 - 8}{h} = 12$')
  plt.grid()
  plt.legend()
  plt.show()

plotarGrafico()
```

# **4. Quarta Questão:**

$$ \lim_{t \to 0} (\frac{1}{t\sqrt{t+1}} - \frac{1}{t}) $$

Substituindo $t$ por $0$, chegamos em:

$$ \lim_{t \to 0} (\frac{1}{0\sqrt{0+1}} - \frac{1}{0}) $$

$$ \lim_{t \to 0} (\frac{1}{0} - \frac{1}{0}) $$

Logo, percebemos que se torna uma divisão por zero, algo que não pode ser feito, então iremos fatorar a equação para chegar no resultado desejado:

$$\lim_{t \to 0}(\frac{\sqrt{t+1}-1}{t\sqrt{t+1}})$$

<br>
Iremos utilizar a regra dos produtos:

$$ \lim_{t \to 0} -\frac{\sqrt{t + 1} - 1} {t \sqrt{t + 1}} = -\lim_{t \to 0} ( \frac{\sqrt{t + 1} - 1}{t}) \lim_{{t \to 0}} (t + 1)^{-1/2}: $$

$$ -1 \left( \lim_{t \to 0} (t + 1)^{-1/2} \right) \left( \lim_{t \to 0} \frac{\sqrt{t + 1} - 1}{t} \right) $$

<br>
Como $\frac{1}{\sqrt{t+1}}$ é continua quando $t = 0$, então $\lim_{t \to 0}\frac{1}{\sqrt{t+1}}$ é igual ao valor da expressão quando $t = 0$, logo:

$$ -\lim_{t \to 0} \frac{\sqrt{t + 1} - 1}{t}$$

Racionalizando a expressão:
$$\frac{{\sqrt{t + 1} - 1}}{t} = \frac{{(\sqrt{t + 1} - 1) (\sqrt{t + 1} + 1)}}{{t (\sqrt{t + 1} + 1)}} = \frac{1}{{\sqrt{t + 1} + 1}}$$


$$ -\lim_{t \to 0} \frac{1}{\sqrt{t+1} + 1}$$

Substituindo o valor de t por 0:
$$ -\lim_{t \to 0} \frac{1}{\sqrt{0+1} + 1} =  \frac{1}{2}$$

Assim chegamos ao resultado:

$$ \lim_{t \to 0} (\frac{1}{t\sqrt{t+1}} - \frac{1}{t}) = -\frac{1}{2} $$

<br>
Fazendo a análise do gráfico abaixo, podemos conferir esse resultado:

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return ((1/ (x * np.sqrt(x+1)))-(1/x))

def plotarGrafico():
  xp = np.linspace(-0.4, 0.4, 100)
  yp = funcao(xp)

  plt.title('Gráfico da função $ \\left(\\frac{1}{t\sqrt{t+1}} - \\frac{1}{t} \\right)$')
  plt.plot(xp, yp, label=r'$f(t)$')
  plt.xlabel('t')
  plt.ylabel('y')
  plt.scatter(0, -0.5, c='red', label='$\\lim_{t \\to 0} \\left(\\frac{1}{t\sqrt{t+1}} - \\frac{1}{t} \\right) = -\\frac{1}{2}}$')
  plt.grid()
  plt.legend()
  plt.show()

plotarGrafico()
```
