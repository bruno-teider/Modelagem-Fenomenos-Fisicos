# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é diretamente proporcional (1:1) a área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.



## **Resposta:**

Nós precisaremos primeiramente calcular a área sobre a curva da função nos intervalos de $x$ equivalendo de $0$ a $2$. Vamos calcular a integral dentro desse intervalo de tempo:

$$ \int_0^2 x^5(2-x) dx $$

Expandindo a função dentro da integral, e integrando cada termo temos:

$$ x^5(2-x) = 2x^5 - x^6 $$

$$ \int_0^2(2x^5-6x^6) dx $$
$$ \int_0^2 2x^5dx - \int_0^2 x^6dx $$

- Cálculo da primeira integral:
$$ \int 2x^5dx = 2 * \frac{x^6}{6} = \frac{2}{6}x^6 =  \frac{1}{3}x^6$$

- Cálculo da segunda integral:
$$ \int x^6 dx = \frac{x^7}{7} $$

Fazendo a junção das diferentes integrais, chegamos na integral definida:
$$ \left[ \frac{1}{3}x^6 - \frac{x^7}{7} \right]_0^2 $$

Analisando ela nos limites:

$$ (\frac{1}{3}(2^6)-\frac{2^7}{7}) - (\frac{1}{3}(0^6) - \frac{0^7}{7}) $$

$$ \frac{64}{3} - \frac{128}{7} $$

Fazendo o MMC chegamos no nosso resultado final:

$$ \frac{64 * 7}{21} - \frac{128*3}{21} = \frac{448}{21} - \frac{384}{21} $$

$$\frac{64}{21} $$

Assim chegamos que a área sobre a curva, e consequentemente a quantidade total de biomassa acumulada nos 2 anos é de $\frac{64}{21}$, ou aproximadamente $3,0476$.

# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate

x = symbols('x')
F = x**5 * (2 - x)

# Cálculo da área sob a curva
def calculoArea():
  area = integrate(F, (x, 0, 2))
  print(f"\nA área sob a curva para x entre 0 e 2 é aproximadamente: {float(area):.4f}\n")

def plotarGrafico():
  area = integrate(F, (x, 0, 2))
  f = lambda x: x**5 * (2 - x) # Converte a função para numpy
  x_vals = np.linspace(0, 2, 100)
  y_vals = f(x_vals)

  plt.figure(figsize=(7, 5))
  plt.plot(x_vals, y_vals, label='$F(x) = x^5(2 - x)$', color='green')
  plt.fill_between(x_vals, y_vals, where=[(x >= 0) and (x <= 2) for x in x_vals], color='skyblue', alpha=0.4, hatch='//', label=f"A = {float(area):.4f}")
  plt.scatter(0, 0, label="x = 0, y = 0", color="red")
  plt.scatter(2, 0, label="x = 2, y = 0", color="blue")

  plt.title('Gráfico da função $F(x) = x^5(2 - x)$ para valores entre 0 e 2')
  plt.xlabel('$x$')
  plt.ylabel('$F(x)$')
  plt.legend()
  plt.grid(True)
  plt.show()

plotarGrafico()
calculoArea()
```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.

## **Resposta:**

Como sabemos que a velocidade é a integral da aceleração, iremos primeiro integrar $a(t)$ para descobrir a equação da velocidade:

$$ v(t) = \int a(t) dt = \int \cos(\pi t) dt $$

Iremos usar a regra da substituição para integrar $cos(πt)$, substituindo $ u = πt $, temos $ du = π dt $ e $ dt = \frac{du}{π} $, e como sabemos que a integral de um $\cos$ é $\sin$:

$$ v(t) = \int \cos(u) \frac{du}{π} = \frac{1}{π} \int \cos(u) du $$

$$ v(t) = \frac{1}{π} \sin(u) + C = \frac{1}{π} \sin(π t) + C $$

Agora substituindo a velocidade no tempo $0$:

$$ v(0) = \frac{1}{π} \sin(0) + C = \frac{1}{2π} $$

$$ C = \frac{1}{2π} $$

Assim chegamos que a equação da velocidade é:

$$ v(t) = \frac{1}{π} \sin(πt) + \frac{1}{2π} $$<br>


Agora que temos a velocidade, iremos encontrar qual a posição. A posição é a integral da velocidade, entao iremos integrar $v(t)$ para os valores de $t$ de $0$ até $t$:

$$ s(t) = \int v(t) dt = \int \left( \frac{1}{π} \sin(π t) + \frac{1}{2π} \right) dt $$

$$ s(t) = \frac{1}{π} \int \sin(π t) dt + \frac{1}{2π} \int 1 dt $$

Vamos utilizar novamente a regra da substituição para resolver a primeira integral:

$$ \int \sin(π t) dt = \int \sin(u) \frac{du}{π} = \frac{1}{π} \int \sin(u) du = -\frac{1}{π} \cos(u) = -\frac{1}{π} \cos(π t) $$

A segunda integral:
$$ \int 1 dt = t $$

Assim temos que a equação da posição no tempo $0$ é:

$$ s(t) = \frac{1}{π} \left( -\frac{1}{π} \cos(π t) \right) + \frac{1}{2π} t + C $$

$$ s(t) = -\frac{1}{π^2} \cos(π t) + \frac{1}{2π} t + C $$

$$ s(0) = -\frac{1}{π^2} \cos(0) + \frac{1}{2π} * 0 + C = 0 $$

$$ -\frac{1}{π^2} + C = 0 $$

$$ C = \frac{1}{π^2} $$

Assim chegamos na equação da posição:

$$ s(t) = -\frac{1}{π^2} \cos(π t) + \frac{1}{2π} t + \frac{1}{π^2} $$<br>

Agora iremos encontrar a distância percorrida nos primeiros $1,5$ segundos ($\cos(1.5\pi) = 0$):

$$ s(1.5) = -\frac{1}{π^2} * 0 + \frac{1}{2π} * 1.5 + \frac{1}{π^2} $$

$$ s(1.5) = \frac{1.5}{2π} + \frac{1}{π^2}$$

$$ s(1.5) = \frac{1.5}{2π} + \frac{1}{π^2} = \frac{0.75}{π} + \frac{1}{π^2} $$

$$ s(1.5) = \frac{1.5}{2π} + \frac{1}{π^2} \approx 0.340054 $$

Assim chegamos na conclusão que a distância percorrida em $1,5$ segundos é de aproximadamente $0.34$ metros.

# Questão 4 - Vale até 1 ponto

Encontre a área limitada por $f(x) = -x^2 + 4x + 1$, $g(x) = -x^3 + 7x^2 - 10x + 3$ para  $1\leq x \leq 2$. Nesta questão você pode usar álgebra, ou Python, considerando as limitações de usos de bibliotecas definidas para esta avaliação.  


```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

def calcArea():
  x = sp.symbols('x')

  # Funções
  f = -x**2 + 4*x + 1
  g = -x**3 + 7*x**2 - 10*x + 3

  diff = f - g

  # Calculo da integral entre 1 e 2
  area = sp.integrate(diff, (x, 1, 2))

  print(f'\nA área limitada das funções é: {float(area):.4f}\n')

def plotarGrafic():
  def f(x):
    return -x**2 + 4*x + 1

  def g(x):
    return -x**3 + 7*x**2 - 10*x + 3

  plt.figure(figsize=(7, 5))
  x = np.linspace(0, 3, 100)
  plt.plot(x, f(x), label='f(x)')
  plt.plot(x, g(x), label='g(x)')

  plt.fill_between(x, f(x), g(x), where=(x >= 1) & (x <= 2), color='skyblue', alpha=0.5, hatch='//',  label='área limitada')

  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Gráfico de $f(x) = -x^2 + 4x + 1$ e $g(x) = -x^3 + 7x^2 -10x + 3$')
  plt.legend()
  plt.grid(True)
  plt.show()

plotarGrafic()
calcArea()
```

# Questão 5 - Vale até 1 ponto

Use o método dos discos para determinar o volume do sólido obtido pela rotação da região delimitada pelas funções $y = \sqrt{x}$, $y = 3$ e o próprio eixo $y$ em torno do eixo $y$.


## **Resposta:**

A primeira coisa que precisamos fazer, é expressar $x$ em função a $y$, como $y = \sqrt{x}$, temos que $ x = y^2 $. <br>

Sabemos que as fronteiras da região em $y$ são de $0$ a $3$. Chegamos no volume do sólido gerado pela rotação em torno do eixo $y$ com a seguinte integral:

$$ V = π \int_{a}^{b} [R(y)]^2 \, dy $$

Onde $R(y)$ é o raio do disco em relação a $y$, que iremos trocar por $y^2$, logo o volume é:

$$ V = π \int_{0}^{3} (y^2)^2 dy $$

$$ V = π \int_{0}^{3} y^4 dy $$

$$ \int_{0}^{3} y^4 dy = \left[ \frac{y^5}{5} \right]_{0}^{3} = \frac{3^5}{5} - \frac{0^5}{5} = \frac{243}{5} $$

$$ V = π * \frac{243}{5} = \frac{243π}{5} $$

$$ V = \frac{243π}{5} $$

Assim chegamos na conclusão que o volume do sólido é:
$$ \frac{243π}{5} ≈ 152.68 $$
