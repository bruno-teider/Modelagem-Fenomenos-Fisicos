## Interpolação de Lagrange usando a biblioteca Scipy.

```python
import numpy as np
from scipy.interpolate import lagrange
from matplotlib import pyplot as plt

#Lista dos pontos
x = np.array([300, 304, 305, 307])
y = np.array([2.4771, 2.4829, 2.4843, 2.4871])
poly = lagrange(x, y)

# Pontos do gráfico
x_plot = np.linspace(min(x), max(x), 100)
y_plot = poly(x_plot)
plt.scatter(x,y, color="red", label="Dados")
plt.grid()

# Plotar o polinomio
plt.plot(x_plot, y_plot, label="Iterpolação Lagrange")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Interpolação de Lagrange")

plt.legend()
plt.show()
```

## Interpolação de Lagrange usando apenas as bibliotecas Numpy e Sympy.

```python
import numpy as np
from sympy import symbols, simplify, lambdify

# Função para a interpolação de lagrange
def interpolacao(x, y):
  n = len(x)
  X = symbols('X')
  poly = 0

  # Realiza a operação da interpolação
  for i in range(n):
    term = y[i]
    for j in range(n):
      if j != i:
        term *= (X - x[j]) / (x[i] - x[j])
    poly += term

  poly = simplify(poly)
  return poly

# Definição das variáveis
x = np.array([300, 304, 305, 307])
y = np.array([2.4771,2.4829,2.4843,2.4871])
poly = interpolacao(x, y)
funcao_lagrange = lambdify('X', poly, 'numpy')

# Plotar o gráfico
x_plot = np.linspace(min(x), max(x), 100)
y_plot = funcao_lagrange(x_plot)
plt.scatter(x,y, color="red", label="Dados")
plt.grid()

plt.plot(x_plot, y_plot, label="Iterpolação de Lagrange")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Interpolação de Lagrange")
plt.legend()
plt.show()
```
