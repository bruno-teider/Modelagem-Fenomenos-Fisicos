Resolva os seguintes limites algebricamente: (a) para as funções tendendo ao infinito; (b) para as funções tendendo a zero. Depois verifique os limites encontrados com o sympy e trace os gráficos das funções.

# f(x) = $ \frac{x-3}{7-x} $

```python
from matplotlib import pyplot as plt
from numpy import linspace
import sympy as sp

def plotarGrafico():
  xp = linspace(-2, 2, 100)
  yp = [(x-3)/(7-x) for x in xp]

  plt.plot(xp, yp, label=r"$f(x) = \frac{x-3}{7-x}$")
  plt.title(r"Gráfico da função $f(x) = \frac{x-3}{7-x}$ pelo MatPlotLib")
  plt.xlabel('X')
  plt.ylabel('f(x)')
  plt.grid()
  plt.legend()
  plt.show()

  # Plotando pelo Sympy
  #y = (x-3)/(7-x)
  #sp.plot(y, (x, -2, 2), title=r'Gráfico da função $f(x) = \frac{x-3}{7-x}$ pelo Sympy')

def calculoLimites():
  x = sp.Symbol('x')
  y = (x-3)/(7-x)

  lim_infinito = sp.limit(y, x, sp.oo)
  lim_zero = sp.limit(y, x, 0)
  print(f"\nLimite da equação para infinito: {lim_infinito} | Limite da equação para zero: {lim_zero} \n")

calculoLimites()
plotarGrafico()
```

# f(x) = $\frac{1}{x^2 + 5x}$

```python
from matplotlib import pyplot as plt
from numpy import linspace
import sympy as sp

def plotarGrafico():
  xp = linspace(1, 100, 100)
  yp = [(1/(x**2 + 5 * x)) for x in xp]

  plt.plot(xp, yp, label=r"$f(x) = \frac{1}{x^2+5x}$")
  plt.title(r"Gráfico da função $f(x) = \frac{1}{x^2+5x}$ pelo MatPlotLib")
  plt.xlabel('X')
  plt.ylabel('f(x)')
  plt.grid()
  plt.legend()
  plt.show()

  # Plotando pelo Sympy
  #y = 1 / ((x ** 2) + (5 * x))
  #sp.plot(y, (x, 1, 100), title=r'Gráfico da função $f(x) = \frac{1}{x^2+5x}$ pelo Sympy')

def calculoLimite():
  x = sp.Symbol("x")
  y = 1 / (x**2 + 5 * x)

  lim_infinito = sp.limit(y, x, sp.oo)
  lim_zero = sp.limit(y, x, 0)
  print(f"\nLimite tendendo ao infinito: {lim_infinito} | Limite tendendo a zero: {lim_zero}\n")

calculoLimite()
plotarGrafico()
```

# $f(x) = \frac{x-1}{x^2 + 5x - 6}$

```python
from matplotlib import pyplot as plt
from numpy import linspace
import sympy as sp

def plotarGrafico():
  xp = linspace(-5.4, 5.4, 100)
  yp = [((x - 1) / ((x ** 2) + (5 * x) - 6)) for x in xp]

  plt.plot(xp, yp, label=r'$f(x) = \frac{x-1}{x^2+5x-6}$')
  plt.title(r'Gráfico da função $f(x) = \frac{x-1}{x^2+5x-6}$ pelo MatPlotLib')
  plt.xlabel('X')
  plt.ylabel('f(x)')
  plt.legend()
  plt.grid()
  plt.show()

  # Plotando pelo Sympy
  #y = (x-1)/(x**2 + (5*x) - 6)
  #sp.plot(y, (x, -5.4, 5.4), title=r'Gráfico da função $f(x) = \frac{x-1}{x^2 + 5x - 6}$ pelo Sympy')

def calcularLimite():
  x = sp.Symbol('x')
  y = (x-1)/(x**2 + (5*x) - 6)

  lim_infinito = sp.limit(y, x, sp.oo)
  lim_zero = sp.limit(y, x, 0)
  print(f"\nLimite tendendo ao infinito: {lim_infinito} | Limite tendendo a zero: {lim_zero} \n")

calcularLimite()
plotarGrafico()
```
