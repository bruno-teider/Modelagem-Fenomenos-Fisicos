# Resposta:
a) $f(x) = x^4 - 4x^3 + 4x^2$
<br><br>

$\frac{d}{dx}4*x^{4-1}-3*4x^{3-1}+2*4x^{2-1}=4x^3-12x^2+8x$

Logo temos que a primeira derivada é: $f'(x) = 4x^3 - 12x^2 + 8x$
<br><br>

Segunda derivada usando regra da potência:
<br>
$\frac{d^2}{dx^2} 3*4x^{3-1} - 2*12x^{2-1} + 1*8x^{1-1} = 12x^2-24x+8$

Logo temos que a segunda derivada é: $f''(x) = 12x^2 - 24x + 8$

<br><br>
Ao igualar a primeira derivada a zero e substituir a, b e c na fórmula quadrática encontramos 3 pontos críticos, sendo eles $x = 0$, $x = 1$ e $x = 2$

Substituindo na segunda derivada $x$ pelos pontos críticos: <br>
$f''(0) = 8$ <br>
$f''(1) = -4$ <br>
$f''(2) = 8$ <br>

Assim, chegamos ao resultado que: <br>
Mínimas Locais: $x = 0$, $x = 2$ <br>

Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

a)$f(x)=e^{2x}−e^x$

b)$f(x)=e^x+x$

# Resposta:

a) $f(x) = e^{2x} - e^x$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def analisePontos():
    x = sp.symbols('x')
    f = sp.exp(2*x) - sp.exp(x)  # Definição da função
    derivada = sp.diff(f, x)  # Acha a derivada
    ponto_critico = sp.solve(derivada, x)  # Encontra os pontos críticos

    print("Primeira derivada:")
    display(derivada.evalf())

    print("\nPontos críticos da função:")
    for ponto in ponto_critico:
        print("x =", ponto.evalf(n=3))

    segunda_derivada = sp.diff(derivada, x)  # Cálculo da segunda derivada

    print("\nSegunda derivada:")
    display(segunda_derivada.evalf())

    print("\nPontos Críticos na segunda derivada:")
    for ponto in ponto_critico:
        valor_segunda_derivada = segunda_derivada.subs(x, ponto)
        print(f"x = {ponto.evalf(n=3)}, a segunda derivada é {valor_segunda_derivada.evalf(n=3)}")

    print("\nAnálise do ponto crítico:")
    for ponto in ponto_critico:
        valor_segunda_derivada = segunda_derivada.subs(x, ponto)
        if valor_segunda_derivada > 0:
            print(f"Sendo x = {ponto.evalf(n=3)}, a segunda derivada é {valor_segunda_derivada.evalf(n=3)}, logo, como o resultado é positivo, é um ponto de mínima.")
        elif valor_segunda_derivada < 0:
            print(f"Sendo x = {ponto.evalf(n=3)}, a segunda derivada é {valor_segunda_derivada.evalf(n=3)}, logo, como o resultado é negativo, é um ponto de máxima.")
        else:
            print(f"Sendo x = {ponto.evalf(n=3)}, a segunda derivada é {valor_segunda_derivada.evalf(n=3)}, e como não é um resultado conclusivo, é um ponto de inflexão.")

def plotarGráfico():
    def f(x): return np.exp(2*x) - np.exp(x)
    def derivada(x): return 2*np.exp(2*x) - np.exp(x)
    def segunda_derivada(x): return 4*np.exp(2*x) - np.exp(x)

    xp = np.linspace(-1.75, 1.75, 100)

    x_ponto = -0.693147180559945
    y_ponto = f(x_ponto)
    y_ponto_derivada = derivada(x_ponto)
    y_ponto_segunda_derivada = segunda_derivada(x_ponto)

    yp = f(xp)
    y_values_derivada = derivada(xp)
    y_values_segunda_derivada = segunda_derivada(xp)

    plt.plot(xp, yp, label='$f(x) = e^{2x} - e^x$', color="#FF4900")
    plt.plot(xp, y_values_derivada, label="Primeira Derivada", color="#011BFF")
    plt.plot(xp, y_values_segunda_derivada, label="Segunda Derivada", color="#3944AA")
    plt.scatter(x_ponto, y_ponto, color='red', label=f'Crítico ({x_ponto:.2f}, {y_ponto:.2f})')
    plt.title('Gráfico de $f(x)$, $f\'(x)$ e $f\'\'(x)$ com Ponto Crítico')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

analisePontos()
print("\n")
plotarGráfico()
```

# Questão 4 - Alternativa - Vale até 2 pontos

Em uma fábrica, latas de sopa cilíndricas devem ser fabricadas para conter um volume dado $V$. Não há desperdício de material na corte da parte vertical de cada lata ( o lado do cilíndro), mas cada tampa e fundo, que são círculos de raio $r$, são cortados de um quadrado que mede $2r$ unidades de lado. Assim, o material usado para fabricar cada lata de sopa tem uma área de $A = 2 \pi rh + 8r^2$ unidades quadradas. Encontre a proporção entre a altura e o diâmetro para a lata mais econômica (ou seja, que requer a menor quantidade de material para fabricação).


# Resposta:

A fórmula para o volume do cilindro é: $V = πr^2h$<br>
Sabemos também que a área da superfície é: $A = 2πrh + 8r^2$<br>

Iremos utilizar a fórmula do volume para conseguir remover $h$ em termos de $d$, e iremos substituir o $r$ por $\frac{d}{2}$, pois o raio é a metade do diâmetro:<br>

$h = \frac{V}{πr^2} = \frac{4V}{πd^2}$ <br>

Logo podemos substituir o $h$ na fórmula da área, onde ficamos com:<br>
$A = 2πr(\frac{4V}{πd^2}) + 8r^2$<br>

$A = \frac{8πV}{d}+8r^2$<br>

Agora iremos utilizar a derivada da área total em relação a $d$ igualando ela a zero, para assim achar o raio:<br>
$-\frac{8πV}{d^2}+16r = 0$<br>
$\frac{8πV}{d^2} = 16r$<br>
$\frac{4πV}{d^2} = r$<br>

Sabemos que $r = \frac{d}{2}$, logo:<br>
$\frac{d}{2} = \frac{4πV}{d^2}$<br>
$d^3 = 8πV$<br>
$d = 2(\frac{V}{\pi})^\frac{1}{3}$<br>

Agora usando a fórmula do volume, iremos encontrar o $h$<br>
$h = \frac{V}{πr^2} = \frac{V}{π(\frac{d}{2})^2}$<br>
$h = \frac{4V}{πd^2} = 2(\frac{V}{π})^\frac{1}{3}$<br>

Logo, chegamos que a proporção entre altura($h$) e diâmetro($d$) para ter o menor disperdício é:<br>
$\frac{h}{d} = \frac{2(\frac{V}{π})^\frac{1}{3}}{2(\frac{V}{π})^\frac{1}{3}} = 1$
